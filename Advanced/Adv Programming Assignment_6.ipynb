### Q1. You are given two strings s and t. String t is generated by randomly shuffling string s and then adding one more letter at a random position. Return the letter that was added to t.

Examples

find_the_difference("abcd", "abcde") ➞ "e"

find_the_difference("", "y") ➞ "y"

find_the_difference("ae", "aea") ➞ "a"


```python
#define function
def find_the_difference(s, t):
    return t[len(s)]

#define fucntion
print('find_the_difference("abcd", "abcde") -> ', find_the_difference("abcd", "abcde"))
print('find_the_difference("", "y") -> ', find_the_difference("", "y"))
print('find_the_difference("ae", "aea") -> ', find_the_difference("ae", "aea"))
```

### Q2. Given a function that accepts unlimited arguments, check and count how many data types are in those arguments. Finally return the total in a list.

List order is:

[int, str, bool, list, tuple, dictionary]

Examples

count_datatypes(1, 45, "Hi", False) ➞ [2, 1, 1, 0, 0, 0]

count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1) ➞ [3, 0, 0, 1, 1, 0]

count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23) ➞ [2, 2, 3, 1, 0, 2]

count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6]) ➞ [2, 0, 1, 2, 2, 0]


```python
#deifne fucntion
def count_datatypes(*args):
    temp_dict = {int: 0, str: 0, bool: 0, list: 0, tuple: 0, dict: 0}
    for arg in args:
        temp_dict[type(arg)] += 1
    
    return list(temp_dict.values())

#testing function
print('count_datatypes(1, 45, "Hi", False) -> ', count_datatypes(1, 45, "Hi", False))
print('count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1) -> ', count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1))
print('count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23) -> ', count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1, 3], {"Brayan": 18}, 25, 23))
print('count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6]) -> ', count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6]))
```

### Q3. A Fibonacci string is a precedence of the Fibonacci series. It works with any two characters of the English alphabet (as opposed to the numbers 0 and 1 in the Fibonacci series) as the initial items and concatenates them together as it progresses in a similar fashion as the Fibonacci series.

Examples

fib_str(3, ["j", "h"]) ➞ "j, h, hj"

fib_str(5, ["e", "a"]) ➞ "e, a, ae, aea, aeaae"

fib_str(6, ["n", "k"]) ➞ "n, k, kn, knk, knkkn, knkknknk"


```python
#define function
def fib_str(n, lst):
    temp_lst = []
    final_str = ""
    first_term = lst[0]
    second_term = lst[1]
    
    
    if n == 0:
        pass
        
    elif n == 1:
        return lst[n - 1]
    
    elif n == 2:
        return lst[n - 1]
        
    elif n > 2:
        temp_lst.append(first_term)
        temp_lst.append(second_term)

        for i in range(2, n):

            first_term = temp_lst[i - 1]

            second_term = temp_lst[i - 2]

            third_term = first_term + second_term

            temp_lst.append(third_term)
    
    final_str = ", ".join(temp_lst)
        
    return final_str

#testing function
print('fib_str(3, ["j", "h"]) ->  ', fib_str(3, ["j", "h"]))
print('fib_str(5, ["e", "a"]) -> ', fib_str(5, ["e", "a"]))
print('fib_str(6, ["n", "k"]) -> ', fib_str(6, ["n", "k"]))
```

### Q5. The Fibonacci sequence is a classic use case for recursive functions since the value of the sequence at a given index is dependent on the sum of the last two values.
